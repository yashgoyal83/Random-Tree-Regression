```python
cats = ['TopicDesc','MeasureDesc','Response','Education','Gender']
df = pd.get_dummies(df, columns=cats, drop_first=True)
```

```python
df.head()
```

```python
x = df.drop(['Tobacco consumption percentage (%)','YEAR','LocationDesc'], axis=1)
y = df['Tobacco consumption percentage (%)']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=101, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=150, random_state=101)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse=mean_squared_error(y_test, pred)
mse
```

```python
mae=mean_absolute_error(y_test, pred)
mae
```
